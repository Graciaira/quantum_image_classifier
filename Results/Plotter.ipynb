```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# save numpy array as csv file
from numpy import asarray
from numpy import savetxt, loadtxt

def save_to_csv(data, file_name):
    # save to csv file
    savetxt(file_name, data, delimiter=',')
    
def load_from_csv(file_name):
    # load from csv file
    return loadtxt(file_name, delimiter=',')
```

```python
label = 0
```

```python
acc_L12_F4 = load_from_csv("C:/Users/user/quantum_image_classifier/PennyLane/Model & Hasil Training/" + str(label) + "/acc_hist_4_features_1.txt")
loss_L12_F4 = load_from_csv("C:/Users/user/quantum_image_classifier/PennyLane/Model & Hasil Training/" + str(label) + "/loss_hist_4_features_1.txt")

acc_L6_F4 = load_from_csv("C:/Users/user/quantum_image_classifier/PennyLane/Model & Hasil Training/" + str(label) + "/acc_hist_4_features_6layer_1.txt")
loss_L6_F4 = load_from_csv("C:/Users/user/quantum_image_classifier/PennyLane/Model & Hasil Training/" + str(label) + "/loss_hist_4_features_6layer_1.txt")

acc_L12_F8 = load_from_csv("C:/Users/user/quantum_image_classifier/PennyLane/Model & Hasil Training/" + str(label) + "/acc_hist_8_features_1.txt")
loss_L12_F8 = load_from_csv("C:/Users/user/quantum_image_classifier/PennyLane/Model & Hasil Training/" + str(label) + "/loss_hist_8_features_1.txt")

acc_L6_F8 = load_from_csv("C:/Users/user/quantum_image_classifier/PennyLane/Model & Hasil Training/" + str(label) + "/acc_hist_8_features_6layer_1.txt")
loss_L6_F8 = load_from_csv("C:/Users/user/quantum_image_classifier/PennyLane/Model & Hasil Training/" + str(label) + "/loss_hist_8_features_6layer_1.txt")
```

```python
iter_history = np.linspace(1, 100, 100, dtype=int)
```

```python
plt.rcParams.update({'font.size': 22})
```

```python
# plot acc train, acc val

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 12
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size

plt.plot(iter_history, acc_L6_F4[:, 0], color='b', label='Latih, k = 4, R = 6', linestyle='-')
plt.plot(iter_history, acc_L6_F4[:, 1], color='g', label='Validasi, k = 4, R = 6', linestyle='-')

plt.plot(iter_history, acc_L12_F4[:, 0], color='r', label='Latih, k = 4, R = 12', linestyle='--')
plt.plot(iter_history, acc_L12_F4[:, 1], color='c', label='Validasi, k = 4, R = 12', linestyle='--')

plt.plot(iter_history, acc_L6_F8[:, 0], color='m', label='Latih, k = 8, R = 6', linestyle='-')
plt.plot(iter_history, acc_L6_F8[:, 1], color='y', label='Validasi, k = 8, R = 6', linestyle='-')

plt.plot(iter_history, acc_L12_F8[:, 0], color='lightcoral', label='Latih, k = 8, R = 12', linestyle='--')
plt.plot(iter_history, acc_L12_F8[:, 1], color='lawngreen', label='Validasi, k = 8, R = 12', linestyle='--')

plt.ylabel('Akurasi')
plt.xlabel('Iterasi/Epoch')
plt.title("Akurasi Latih dan Validasi Label " + str(label) + " Lawan Semua")

plt.legend()
plt.show()
```

```python
# plot loss train, loss val

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 12
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size

plt.plot(iter_history, loss_L6_F4[:, 0], color='b', label='Latih, k = 4, R = 6', linestyle='-')
plt.plot(iter_history, loss_L6_F4[:, 1], color='g', label='Validasi, k = 4, R = 6', linestyle='-')

plt.plot(iter_history, loss_L12_F4[:, 0], color='r', label='Latih, k = 4, R = 12', linestyle='--')
plt.plot(iter_history, loss_L12_F4[:, 1], color='c', label='Validasi, k = 4, R = 12', linestyle='--')

plt.plot(iter_history, loss_L6_F8[:, 0], color='m', label='Latih, k = 8, R = 6', linestyle='-')
plt.plot(iter_history, loss_L6_F8[:, 1], color='y', label='Validasi, k = 8, R = 6', linestyle='-')

plt.plot(iter_history, loss_L12_F8[:, 0], color='lightcoral', label='Latih, k = 8, R = 12', linestyle='--')
plt.plot(iter_history, loss_L12_F8[:, 1], color='lawngreen', label='Validasi, k = 8, R = 12', linestyle='--')

plt.ylabel('Nilai Biaya Total')
plt.xlabel('Iterasi/Epoch')
plt.title("Nilai Biaya Total Latih dan Validasi Label " + str(label) + " Lawan Semua")

plt.legend()
plt.show()
```
